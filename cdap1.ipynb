```python
#top 10 song whose popularity is grater than 90
most_popular=df.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10] 
```

```python
df.set_index("release_date", inplace=True)
df.index=pd.to_datetime(df.index)
df.head() 
```

```python
df[["artists"]].iloc[18] #artist present at 18th row location
```

```python
#duration in ms to sec 
df["duration"]=df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms", inplace=True, axis=1)
```

```python
df.duration.head()
```

```python
#visualisation ie correlation map
corr_df=df.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
#-1.00 negative correlation /least and values above 0.0 has positive correaltion
```

```python
sample_df=df.sample(int(0.004*len(df)))#.4% of dataset 
```

```python
print(len(sample_df))
```

```python
#regression plot bet loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness", x="energy", color="c").set(title="Loudness vs energy correation")
#energy increases the loudness of the song increases and if loudness of song decreases then energy also get decreased 
```

```python
#popularity vs acousticness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity", x="acousticness", color="b").set(title="popularity vs acousticness correlation")
```

```python
df['dates']=df.index.get_level_values('release_date')
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
```

```python
sns.displot(years,discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_dr=df.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years, y= total_dr, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)#bar plot for duration and year in sec 
```

```python
total_dr=df.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)#Line plot 
```
